Алгоритм Introselect изобретён Дейвом Массером (более известен как один из "отцов" Standard Template Library и парадигмы generic programming).

```python
import numpy as np
```

```python
relevance = np.random.rand(10000000)
print(relevance)
```

```python
%%time
top_indices = (-relevance).argsort()[:100]
top_relevance = relevance[top_indices]
recs1 = [
    (pair[0], pair[1])
    for pair in zip(top_indices, top_relevance)
]
```

```python
%%time
top_indices = (-relevance).argpartition(100)[:100]
top_relevance = relevance[top_indices]
recs2 = sorted(
    [
        (pair[0], pair[1])
        for pair in zip(top_indices, top_relevance)
    ],
    key=lambda pair: -pair[1]
)
```

```python
print(recs1 == recs2)
print(recs1)
```
